Lakukan pra-pemrosesan awal seperti imputasi nilai hilang untuk kolom `TransactionDT`, konversi ke format datetime, dan rekayasa fitur temporal (jam, hari dalam seminggu, hari dalam setahun). Juga, buat fitur agregat untuk 'card1' dan 'card4' berdasarkan 'TransactionAmt'.

## Identify Column Types and Handle Missing Values

### Subtask:
Identifikasi kembali kolom numerik dan kategorikal setelah rekayasa fitur. Imputasi nilai hilang di kolom numerik menggunakan median/mean dan di kolom kategorikal dengan 'Missing'.


**Reasoning**:
The first instruction is to separate columns into numerical and categorical lists for both `df_train` and `df_test`, excluding 'TransactionID' and 'isFraud'. I will iterate through the columns and assign them based on their data types.



```python
print("Separating columns into numerical and categorical lists...")

# Identify numerical and categorical columns for df_train
numerical_cols_train = df_train.select_dtypes(include=['int64', 'float64', 'int32']).columns.tolist()
categorical_cols_train = df_train.select_dtypes(include=['object']).columns.tolist()

# Remove 'TransactionID' and 'isFraud' from the lists if present
if 'TransactionID' in numerical_cols_train:
    numerical_cols_train.remove('TransactionID')
if 'isFraud' in numerical_cols_train:
    numerical_cols_train.remove('isFraud')

# Identify numerical and categorical columns for df_test
numerical_cols_test = df_test.select_dtypes(include=['int64', 'float64', 'int32']).columns.tolist()
categorical_cols_test = df_test.select_dtypes(include=['object']).columns.tolist()

# Remove 'TransactionID' from the lists if present
if 'TransactionID' in numerical_cols_test:
    numerical_cols_test.remove('TransactionID')

print(f"Numerical columns (train): {len(numerical_cols_train)}")
print(f"Categorical columns (train): {len(categorical_cols_train)}")
print(f"Numerical columns (test): {len(numerical_cols_test)}")
print(f"Categorical columns (test): {len(categorical_cols_test)}")
```

**Reasoning**:
Now that the numerical and categorical columns have been identified, I will proceed with imputing missing values in the numerical columns. As per the instructions, I will calculate the median for each numerical column from `df_train` and use it to fill missing values in both `df_train` and `df_test` for that column.



```python
print("Imputing missing values in numerical columns...")

for col in numerical_cols_train:
    if df_train[col].isnull().any():
        median_val = df_train[col].median()
        df_train[col].fillna(median_val, inplace=True)
        df_test[col].fillna(median_val, inplace=True)
print("Missing values in numerical columns imputed using median from df_train for both dataframes.")
```

**Reasoning**:
The previous step successfully imputed numerical missing values, but a `FutureWarning` was raised about `inplace=True`. While not an error, it's good practice to update the code to avoid future issues. The next instruction is to impute missing values in categorical columns. I will iterate through the identified categorical columns and fill `NaN` values with the string 'Missing' for both `df_train` and `df_test`.



```python
print("Imputing missing values in categorical columns...")

for col in categorical_cols_train:
    if df_train[col].isnull().any():
        df_train[col] = df_train[col].fillna('Missing')
    if df_test[col].isnull().any():
        df_test[col] = df_test[col].fillna('Missing')
print("Missing values in categorical columns imputed with 'Missing' for both dataframes.")
```

**Reasoning**:
Now that both numerical and categorical missing values have been imputed, I need to verify that there are no remaining missing values in either `df_train` or `df_test` by printing the sum of nulls for both dataframes, as per the instructions.



```python
print("Verifying no remaining missing values...")

missing_train = df_train.isnull().sum().sum()
missing_test = df_test.isnull().sum().sum()

print(f"Total missing values in df_train: {missing_train}")
print(f"Total missing values in df_test: {missing_test}")

if missing_train == 0 and missing_test == 0:
    print("No missing values found in df_train and df_test after imputation.")
else:
    print("WARNING: Missing values still exist in df_train or df_test.")
```

The previous verification step indicated that `df_test` still has 550 missing values. This will help understand if they are columns that were supposed to be imputed or if there's an issue with the column identification.



```python
print("Identifying columns in df_test with remaining missing values...")

missing_cols_df_test = df_test.isnull().sum()
missing_cols_df_test = missing_cols_df_test[missing_cols_df_test > 0]

if not missing_cols_df_test.empty:
    print("Columns in df_test with remaining missing values and their counts:")
    print(missing_cols_df_test)
else:
    print("No missing values found in df_test.")
```

The previous numerical imputation step missed filling missing values in 'D1' of `df_test` because the imputation logic was conditional on `df_train` having missing values in that column. To properly address all missing numerical values as per the instructions (which state to use the median from `df_train` to fill `NaN`s in both `df_train` and `df_test`), I will re-execute the numerical imputation step, ensuring that the median from `df_train` is calculated for each numerical column and then used to fill missing values in *both* `df_train` and `df_test`, regardless of whether `df_train` itself has missing values for that column. This will ensure 'D1' in `df_test` is also handled.

Skalakan fitur numerik pada `X_resampled` dan `X_val` menggunakan `MinMaxScaler` untuk mempersiapkan data bagi model deep learning.


**Reasoning**:
The first instruction is to import the `MinMaxScaler` from `sklearn.preprocessing` to prepare for scaling the numerical features.



```python
from sklearn.preprocessing import MinMaxScaler
print("MinMaxScaler imported successfully.")
```

**Reasoning**:
As per the instructions, I will initialize a `MinMaxScaler` object. It's good practice to assign it to a variable, usually named `scaler`.



```python
print("Initializing MinMaxScaler object...")

scaler = MinMaxScaler()

print("MinMaxScaler object initialized successfully.")
```

**Reasoning**:
As per instruction 3, I will now fit the `MinMaxScaler` on the numerical columns of `X_resampled` and then transform these columns, storing the result in `X_resampled_scaled`. I will also transform the corresponding numerical columns in `X_val` using the same fitted scaler, storing the result in `X_val_scaled`, as per instruction 4. I need to explicitly select the numerical columns that are not boolean for scaling.



```python
print("Fitting scaler on X_resampled and transforming X_resampled and X_val...")

# Identify numerical columns in X_resampled that are not boolean for scaling
numerical_cols_for_scaling = X_resampled.select_dtypes(include=['int64', 'float64', 'int32']).columns.tolist()

# Fit the scaler on the numerical columns of X_resampled and transform
X_resampled_scaled_array = scaler.fit_transform(X_resampled[numerical_cols_for_scaling])

# Transform the numerical columns of X_val using the same fitted scaler
X_val_scaled_array = scaler.transform(X_val[numerical_cols_for_scaling])

# Create new DataFrames with scaled numerical columns, keeping other columns (e.g., boolean) as they are
# First, copy the original DataFrames
X_resampled_scaled = X_resampled.copy()
X_val_scaled = X_val.copy()

# Then update the numerical columns with their scaled versions
X_resampled_scaled[numerical_cols_for_scaling] = X_resampled_scaled_array
X_val_scaled[numerical_cols_for_scaling] = X_val_scaled_array

print("X_resampled_scaled and X_val_scaled created with numerical features scaled.")
```

**Reasoning**:
As per instruction 5, I will print the shapes of `X_resampled_scaled` and `X_val_scaled` to verify that the scaling and transformation were applied correctly and to confirm the dimensions of the resulting datasets.



```python
print("Verifying shapes of scaled datasets:")
print(f"X_resampled_scaled shape: {X_resampled_scaled.shape}")
print(f"X_val_scaled shape: {X_val_scaled.shape}")
```

**Reasoning**:
As per instruction 6, I will now display the first 5 rows of `X_resampled_scaled` and `X_val_scaled` using `.head()` to inspect the scaled data.



```python
print("First 5 rows of X_resampled_scaled:")
display(X_resampled_scaled.head())

print("\nFirst 5 rows of X_val_scaled:")
display(X_val_scaled.head())
```

## Train Deep Learning Model (Classification)

### Subtask:
Bangun dan latih model neural network sederhana menggunakan TensorFlow/Keras untuk tugas klasifikasi biner. Gunakan `X_resampled` yang sudah diskalakan dan `y_resampled` untuk pelatihan.


**Reasoning**:
The first instruction is to import the necessary libraries, `tensorflow` and modules like `Sequential` and `Dense` from `keras`, to begin building the deep learning model.



```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

print("TensorFlow and Keras modules imported successfully.")
```

**Reasoning**:
As per instruction 2, I will initialize a Sequential model. This is the foundation for building a layered neural network in Keras.



```python
print("Initializing the Sequential model...")

model_dl = Sequential()

print("Sequential model initialized successfully.")
```

**Reasoning**:
As per instruction 3, I will add the input layer to the sequential model. This layer will be a `Dense` layer with a specified number of units (e.g., 128), `relu` activation, and `input_dim` set to the number of features in `X_resampled_scaled`.



```python
print("Adding input layer to the model...")

# Get the number of features from X_resampled_scaled
input_dim = X_resampled_scaled.shape[1]

model_dl.add(Dense(128, activation='relu', input_dim=input_dim))

print("Input layer added successfully.")
```

**Reasoning**:
Following the instructions, I will now add a hidden layer to the neural network model. This will be a `Dense` layer with a `relu` activation function, building upon the input layer already defined.



```python
print("Adding hidden layer to the model...")

model_dl.add(Dense(64, activation='relu')) # Adding a hidden layer with 64 units

print("Hidden layer added successfully.")
```

**Reasoning**:
As per instruction 5, I will now add the output layer to the model. For binary classification, this layer will be a `Dense` layer with 1 unit and a `sigmoid` activation function.



```python
print("Adding output layer to the model...")

model_dl.add(Dense(1, activation='sigmoid')) # Output layer for binary classification

print("Output layer added successfully.")
```

**Reasoning**:
As per instruction 6, I will now compile the deep learning model. This involves specifying the `Adam` optimizer, `binary_crossentropy` as the loss function (suitable for binary classification), and `accuracy` as the metric to monitor during training.



```python
print("Compiling the deep learning model...")

model_dl.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Model compiled successfully.")
```

**Reasoning**:
As per instruction 7, I will now train the compiled deep learning model using `X_resampled_scaled` and `y_resampled` for training, and `(X_val_scaled, y_val)` for validation, with specified epochs and batch size. This will initiate the learning process.



```python
print("Training the deep learning model...")

history = model_dl.fit(X_resampled_scaled, y_resampled,
                        epochs=10,
                        batch_size=32,
                        validation_data=(X_val_scaled, y_val),
                        verbose=1)

print("Deep learning model trained successfully.")
```
